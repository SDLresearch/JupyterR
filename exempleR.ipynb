```r
data(meuse)
str(meuse)
```

```r
coordinates(meuse) = ~x+y
plot(meuse)
```

```r
vario = variogram(log(zinc)~1, meuse)
plot(vario)
```

```r
vario.fit = fit.variogram(vario, model = vgm(psill = 1, model = 'Sph', range = 900, nugget = 0.5))
plot(vario, vario.fit)
```
